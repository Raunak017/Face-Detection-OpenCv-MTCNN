# Face Detection Using MTCNN

```python
# face detection with mtcnn on a photograph
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from mtcnn.mtcnn import MTCNN

# load image from file
sample = 'footballTeam.jpg'
sample = pyplot.imread(sample)
pyplot.imshow(sample)
```

```python
# draw an image with detected objects
def draw_image_with_boxes(filename, result_list):
    # load the image
    #data = pyplot.imread(filename)
    # plot the image
    pyplot.imshow(filename)
    # get the context for drawing boxes
    ax = pyplot.gca()
    # plot each box
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = Rectangle((x, y), width, height, fill=False, color='green')
        
        # draw the box
        ax.add_patch(rect)
    
    # show the plot
    pyplot.show()
```

```python
# create the detector, using default weights
detector = MTCNN()
```

```python
# detect faces in the image
faces = detector.detect_faces(sample)
print('Faces Detected:', len(faces))

# display faces on the original image
draw_image_with_boxes(sample, faces)
```

```python
test1 = 'test1.jpg'
test1 = pyplot.imread(test1)
pyplot.imshow(test1)

af1 = 15 # Number of actual faces
print('Actual Faces:', af1)

# detect faces in the image
faces = detector.detect_faces(test1)
print('Faces Detected:', len(faces))

# display faces on the original image
draw_image_with_boxes(test1, faces)
```

We can see that the model predicted 3 False Positives in this test

```python
test2 = 'test2.jpg'
test2 = pyplot.imread(test2)
pyplot.imshow(test2)

af2 = 15 # Number of actual faces
print('Actual Faces:', af2)

# detect faces in the image
faces = detector.detect_faces(test2)
print('Faces Detected:', len(faces))

# display faces on the original image
draw_image_with_boxes(test2, faces)
```

The model detected each face with 100% accuracy in this case
